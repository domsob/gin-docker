### Profile the Project:

```
%%bash
cd /opt/jcodec
projectnameforgin="jcodec"
java -Dtinylog.level=trace \
    -cp /opt/gin/build/gin.jar \
    gin.util.Profiler \
    -r 20 \
    -mavenHome /root/.sdkman/candidates/maven/current \
    -p $projectnameforgin \
    -d . \
    -o $projectnameforgin.Profiler_output.csv
```

### Masking Random Search:

```
%%bash
cd /opt/jcodec
mkdir -p results
projectnameforgin="jcodec"; LLM="gemma2:2b"; model=$LLM
java -Dtinylog.level=trace \
    -cp /opt/gin/build/gin.jar \
    gin.util.RandomSampler \
    -j \
    -p $projectnameforgin \
    -d . \
    -m $projectnameforgin.Profiler_output.csv \
    -o results/$projectnameforgin.RandomSampler_1000_output.$model.csv \
    -mavenHome /root/.sdkman/candidates/maven/current \
    -timeoutMS 10000 \
    -et gin.edit.llm.LLMMaskedStatement \
    -mt $model \
    -pt MASKED \
    -pn 1000 &> results/$projectnameforgin.RandomSampler_COMBINED_1000_stderrstdout.$model.txt
```

### Masking Local Search:

```
%%bash
cd /opt/jcodec
mkdir -p results
projectnameforgin="jcodec"; LLM="gemma2:2b"; model=$LLM
java -Dtinylog.level=trace \
    -cp /opt/gin/build/gin.jar \
    gin.util.LocalSearchRuntime \
    -j \
    -p $projectnameforgin \
    -d . \
    -m $projectnameforgin.Profiler_output.csv \
    -o results/$projectnameforgin.LocalSearchRuntime_COMBINED_50_output.$model.csv \
    -mavenHome /root/.sdkman/candidates/maven/current \
    -timeoutMS 10000 \
    -et gin.edit.llm.LLMMaskedStatement \
    -mt $LLM \
    -pt MASKED \
    -in 100 &> results/$projectnameforgin.LocalSearchRuntime_LLM_MASKED_50_stderrstdout.$model.txt
```
